CREATING COLUMN

```python
nba['selisih_poin'] = nba ['pts'] -  nba ['opp_pts']
nba.head()
```

```python
# Siapkan list kosong. untuk tempat penampung
# specify variabel yang dibutuhkan
# buat prosedur --> hasil
# hasilnya simpan ke list kosong
# loop selesai, list penuh --> kolom baru

hasil = []

# kolom pts
# kolom opp_pts
for i in range(0, len(nba)):
    selisih = nba.loc[i,'pts'] - nba.loc[i,'opp_pts']
    hasil.append(selisih)

hasil
nba['selisih_loop']=hasil
nba.head()

    

```

```python
range(0, len(nba))
```

```python
# kolom baru hasil pertandingan
# def dunction
def hasil(selisih):
    if selisih == 0:
        return 'draw'
    elif selisih > 0:
        return 'menang'
    else:
        return 'kalah'
    
```

```python
hasil_pertandingan = []

for i in range (0,len(nba)):
    result = hasil(nba.iloc[i,-2]) 
    hasil_pertandingan.append(result)

nba['hasil'] = hasil_pertandingan
```
